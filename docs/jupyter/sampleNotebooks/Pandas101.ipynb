```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()  ## cumulative sum

df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index,columns=['A', 'B', 'C', 'D'])
df = df.cumsum()


df.to_csv('foo.csv')
pd.read_csv('foo.csv')
df
```
